```python
import os
import numpy as np
import pandas as pd

from tqdm.auto import tqdm
```

```python
data_2_dir_path = 'data\\ver_2'
data_2_modify_path = 'data\\ver_2_modify'
rpm_index_list = [900, 1200, 1500]
data_2_file_path_list = []

for file_path in os.listdir(data_2_dir_path):
    data_2_file_path_list.append(os.path.join(data_2_dir_path, file_path))
```

```python
data_2_file_path_list
```

```python
raw_data = pd.read_csv(data_2_file_path_list[0], encoding='cp949')
raw_data
```

```python
col_name = raw_data.iloc[6].values
col_name
```

```python
col_name = np.insert(col_name, 0, '시간 [s]')
col_name
```

```python
data_value = raw_data.iloc[7:, :].values
data_value = data_value.astype(np.float64)

data_value
```

```python
time_arr = np.arange(0, data_value.shape[0])/100
time_arr
```

```python
data_value = np.insert(data_value, 0, time_arr, axis=1)
data_value
```

```python
data_pd = pd.DataFrame(data_value, columns=col_name)
data_pd
```

```python
filtered_data = data_pd[data_pd['회전수 [rpm]'] >= rpm_index_list[0]-rpm_index_list[0]*0.05]
filtered_data = filtered_data[filtered_data['회전수 [rpm]'] <= rpm_index_list[0]+rpm_index_list[0]*0.05]
```

```python
filtered_data
```

```python
filtered_data.to_csv('filtered_data.csv',  encoding='cp949', index=False)
```

```python
total_iter =  len(data_2_file_path_list) * len(rpm_index_list)

with tqdm(total=total_iter, desc='Data Splitting...') as pbar:
    for data_2_file_path in data_2_file_path_list:
        for rpm_index in rpm_index_list:
            raw_data = pd.read_csv(data_2_file_path, encoding='cp949')
        
            col_name = raw_data.iloc[6].values
            col_name = np.insert(col_name, 0, '시간 [s]')
        
            data_value = raw_data.iloc[7:, :].values
            data_value = data_value.astype(np.float64)
            
            time_arr = np.arange(0, data_value.shape[0])/100
            data_value = np.insert(data_value, 0, time_arr, axis=1)
            
            data_pd = pd.DataFrame(data_value, columns=col_name)
            
            filtered_data = data_pd[data_pd['회전수 [rpm]'] >= (rpm_index-(rpm_index*0.05))]
            filtered_data = filtered_data[filtered_data['회전수 [rpm]'] <= (rpm_index+(rpm_index*0.05))]
            
            file_name, file_format = os.path.splitext(os.path.basename(data_2_file_path))
            new_file_name = os.path.join(data_2_modify_path, file_name + '_' + str(rpm_index) + file_format)

            filtered_data.to_csv(new_file_name, encoding='cp949', index=False)
            pbar.update(1)
```
